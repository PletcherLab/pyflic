# Load Experiment
Loads the experiment from `python_test_data/` and displays basic results.

```python
import sys
from pathlib import Path

# Find the pyflic repo root by walking up until we find pyproject.toml.
def _find_repo_root(marker: str = "pyproject.toml") -> Path:
    for p in [Path().resolve(), *Path().resolve().parents]:
        if (p / marker).exists():
            return p
    raise RuntimeError(f"Could not find repo root (no {marker} found in any parent directory).")

repo_root = _find_repo_root()

# The parent of repo_root (e.g. GithubLocal/) must be on sys.path so that
# 'import pyflic' resolves to the pyflic/ directory.
if str(repo_root.parent) not in sys.path:
    sys.path.insert(0, str(repo_root.parent))

from pyflic import Experiment
```

```python
project_directory = repo_root / "python_test_data"

exp = Experiment.load(project_directory)
print(exp.summary_text())
```
